```python
'''Separate the dependent and the independent variables'''

iris_data_d = iris_data ['species']
iris_data_i = iris_data [['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# See that names of columns are within two square brackets
```

```python
'''Now to see the count of data in the dependent variable, as to see how are they divided among themselves'''

print (iris_data_d.unique(), '\n')
# Gives the unique items from the column

print (iris_data_d.value_counts())

# The counts are 50 each, meaning the data is balanced. Otherwise, when data is imbalanced, then other approach is used than below methods to analyse the data
```

```python
# 2-D Scatter Plot
#Plotting speal length and petal width

iris_data_i.plot (kind = 'scatter', x = 'sepal_length', y = 'sepal_width', color = 'r', label = 'sepal')
plt.title('Sepal Relation')
plt.grid()
plt.legend()

plt.show

#It tells that Sepal length varies between 4.2 and 8 and width varies between 2 and 4.5
```

```python
# Now we need to see the plot as per different speciesWe to get insights from the scatter plot. We will use seaborn for this

sns.set_style ('whitegrid')
sns.FacetGrid (iris_data, hue = 'species', height = 5)\
.map (plt.scatter, 'sepal_length', 'sepal_width')\
.add_legend()

plt.show()

# From this, it can be concluded that
# 1) The species setosa can be easily separated from the other two species, by drawing a line
# 2) The versicolor and virginica cannot be separated easily, becuase they overlap considerably

# In the syntax above '\' is used after the lines. It is the symbol for line conitnuation character, meaning the next line is in continuation with the above line
```

```python
# 3-D Scatter Plot
#For this plot.ly is used to visualize the 3D scatter plot

#Humans can only visaulize 3D image via eyes, but not beyond 3D. For that the trick used is pair-plot
```

```python
#Pair Plot
# Will have nC2 plots unique plots

sns.set_style ('whitegrid')
sns.pairplot (iris_data, hue = 'species', height = 3)
plt.show()

# In this, there are 12 plots, but of which 6 are just the mirror of the other 6
# So we will just focus on the 6 graphs above the diagonal
```

```python
# In the above pair plot, some easy conslucsions:
# 1) The combination of PL and PW gives the best combination to segregate the species using these two features combination
#    So we can simply use the plot combination of PL and PW to identify any given species type with their PL & PW
# 2) Even by using linear line to segregate and use if else command, the segregation can be done easily.
```
